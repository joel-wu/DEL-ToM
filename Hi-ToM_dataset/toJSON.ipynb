```python
import json
import os
import itertools

input_folder = 'Hi-ToM_data'
lengths = [1, 2, 3]
orders = [0, 1, 2, 3, 4]
prompts = ['CoT', 'MC']
tells = ['No_Tell', 'Tell']

prompt_types = {'CoT': 'CoTP', 'MC': 'VP'}
deception = {'No_Tell': False, 'Tell': True}

# # Directory containing text files
# text_dir = 'path/to/text_files/'

# Output JSON file
output_json = 'Hi-ToM_data.json'

# Creating a dictionary to store filename: content pairs
data = {'data': []}

sample_idx = 0

# Iterate through text files
for tell, prompt, length, order, sample_num in itertools.product(tells, prompts, lengths, orders, range(1, 21)):
    file_path = os.path.join(input_folder, tell, prompt, f'length_{length}', f'sample_{sample_num}',
                            f'order_{order}.txt')
    # Read text content
    new_lines = []
    with open(file_path, 'r', encoding='utf-8') as text_file:
        lines = text_file.readlines()
        print(lines)
        for line in lines:
            if 'Provide' in line.split(' '):
                continue
            elif line == 'Story:\n':
                continue
            elif 'Question:' in line.split(' '):
                question = line.split(' ', 1)[1].split('\n')[0]
            elif 'Answer:' in line.split(' '):
                correct_answer = line.split(' ', 1)[1].split('\n')[0]
            elif 'Choices:' in line.split(' '):
                choices = line.split(' ', 1)[1].split('\n')[0]
            elif 'Note:' in line.split(' '):
                continue
            else:
                new_lines.append(line)
        #content = text_file.read()
        content = ''.join(new_lines)
        full_prompt = 'Read the following story and answer the multiple-choice question. Think step-by-step. Provide the answer first, and then explain it.\nStory:\n' + content + f'\nQuestion: {question}\nChoices: {choices}\n\n' + 'Note: You should assume the following. (1) An agent witnesses everything and every movement before exiting a location. (2) An agent A can infer another agent B\'s mental state only if A and B have been in the same location, or have private or public interactions. (3) Note that every agent tends to lie. What an agent A tells others doesn\'t affect A\'s actual belief. An agent tends to trust an agent that exited the room later than himself. The exit order is known to all agents. (4) Agents in private communications know that others won\'t hear them, but they know that anyone can hear any public claims.\n'
        
        # Add filename and content to the dictionary
        data['data'].append({'prompting_type': prompt_types[prompt], 'deception': deception[tell], 'story_length': length, 'question_order': order, 'sample_id': sample_idx, 'story': content, 'question': question, 'choices': choices, 'answer': correct_answer, 'prompt': full_prompt})
        sample_idx += 1

# Write to JSON
with open(output_json, 'w', encoding='utf-8') as file:
    json.dump(data, file, ensure_ascii=False, indent=4)



```
